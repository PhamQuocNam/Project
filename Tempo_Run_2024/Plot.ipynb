```python
import matplotlib.pyplot as plt
import cv2
from PIL import Image
from torchvision.transforms import ToTensor
```

```python
img_path='train_modified(2)/images/IMG_0132.jpg'
image= Image.open(img_path).convert("RGB")
transform= ToTensor()
image= transform(image)
```

```python
sample=image.permute(1,2,0).numpy().copy()
print(sample.shape)
sample = cv2.rectangle(sample,(0,760),(390, 850),(255,0,0), 3)
plt.imshow(sample)
```

```python
def xywh(x_min,y_min,x_max,y_max,image_width,image_height):
    w= x_max-x_min
    h= y_max-y_min
    x = ((x_max+x_min)/2)/image_width
    y = ((y_max+y_min)/2)/image_height
    w /= image_width
    h /= image_height
    
    return x,y,w,h
```

```python
print(xywh(0,760,390, 850,1125, 1365))
```
